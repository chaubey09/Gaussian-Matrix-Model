```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('GOOGL.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df = df[['Open', 'High', 'Low', 'Close', 'Volume']].dropna()
```

```python
print('Anmol Chaubey 24MAI0111')
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(df)
```

```python
gmm = GaussianMixture(n_components=3, random_state=0)
gmm.fit(X)
```

```python
clusters = gmm.predict(X)
```

```python
df['Cluster'] = clusters
```

```python
print(df.head())
```

```python
print('Anmol Chaubey 24MAI0111')
```

```python
plt.figure(figsize=(12, 8))

# Plot stock prices
plt.plot(df.index, df['Close'], label='Close Price', color='black', alpha=0.5)

# Plot cluster annotations
for cluster in sorted(df['Cluster'].unique()):
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(cluster_data.index, cluster_data['Close'], label=f'Cluster {cluster}', alpha=0.7)

plt.title('Stock Price using GMM cluster by 24MAI0111')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
